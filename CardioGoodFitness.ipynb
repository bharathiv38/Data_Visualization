## Data Visualization

```python
sns.set_theme()
sns.pairplot(data=treadmill)
```

```python
treadmill['Gender'].value_counts()
```

```python
Genderdist = treadmill.groupby(['Product','Gender'])['Gender'].count()
Totallist = treadmill.groupby('Product')['Gender'].count()
Gender_list = round(Genderdist/Totallist*100)
print(Gender_list)
```

```python
sns.countplot(x='Gender',hue='MaritalStatus',data=treadmill)
```

```python
sns.countplot(x='Product',hue='MaritalStatus',data=treadmill)
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='Age',y='Income',hue='Gender',data=treadmill,size='Miles')
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='Usage',y='Income',hue='Product',data=treadmill,size='Miles')
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='Age',y='Miles',data=treadmill,hue='Gender',size='Usage')
```

```python
plt.figure(figsize=(8,8))
sns.swarmplot(x='Fitness',y='Miles',data=treadmill,hue='Gender')
```
