```python
#create 2 new columns
epf_df['MinsPerMatch'] = (epf_df['Mins']/epf_df['Matches']).astype(int)
epf_df['GoalsPerMatch'] = (epf_df['Goals']/epf_df['Matches']).astype(float)
```

```python
print(epf_df['MinsPerMatch'])
```

```python
epf_df.head()
```

```python
#total goals
Total_Goals = epf_df['Goals'].sum()
print(Total_Goals)
```

```python
#penalty goals
Total_Penalty = epf_df['Penalty_Goals'].sum()
print(Total_Penalty)
```

```python
#Penalty Attempts
Total_Penalty_Attempts = epf_df['Penalty_Attempted'].sum()
print(Total_Penalty_Attempts)
```

```python
#Pie chart for penalties missed vs score
plt.figure(figsize=(13,6))
pl_not_scored = epf_df['Penalty_Attempted'].sum() - Total_Penalty
data = [pl_not_scored, Total_Penalty]
labels = ['Penalties Missed','Penalties Scored']
color = sns.color_palette('Set2')
plt.pie(data,labels=labels,colors=color,autopct= '%.0f%%')
plt.show()
```

```python
#most of the players from which countries
nationality = epf_df.groupby('Nationality').size().sort_values(ascending=False)
nationality.head(10).plot(kind='bar', figsize=(12,6),color=sns.color_palette("magma"))
```

```python
#clubs with maximum playersin their squad
epf_df['Club'].value_counts().nlargest(5).plot(kind='bar', color=sns.color_palette('viridis'))
```

```python
#Clubs with the least palyers in the squad
epf_df['Club'].value_counts().nsmallest(5).plot(kind='bar', color=sns.color_palette('viridis'))
```

```python
#players based on age group
under20 = epf_df[epf_df['Age'] <= 20]
under20_25 = epf_df[(epf_df['Age'] > 20) & epf_df['Age'] <= 25]
under25_30 = epf_df[(epf_df['Age'] > 25) & epf_df['Age'] <= 30]
above30 = epf_df[epf_df['Age'] > 30]
```

```python
x = np.array([under20['Name'].count(),under20_25['Name'].count(),under25_30['Name'].count(),above30['Name'].count()])
mylabels=["<=20",">20 & <=25",">25 & <=30",">30"]
plt.title("Total Players with Age",fontsize= 20)
plt.pie(x, labels = mylabels,autopct="%.1f%%")
plt.show()
```

```python
#total under 20 players in each club
players_under_20 = epf_df[epf_df["Age"] < 20]
players_under_20['Club'].value_counts().plot(kind='bar',color=sns.color_palette("cubehelix"))
```

```python
#under 20 play in MancU
players_under_20[players_under_20['Club']=='Manchester United']
```

```python
players_under_20[players_under_20['Club']=='Chelsea']
```

```python
#Average player in each club
plt.figure(figsize=(20,10))
sns.boxplot(x ='Club',y='Age',data=epf_df)
plt.xticks(rotation=90)
```

```python
num_player = epf_df.groupby('Club').size()
data=(epf_df.groupby('Club')['Age'].sum())/num_player
data.sort_values(ascending=False)
```

```python
# Total assist from each club

Assists_by_clubs = pd.DataFrame(epf_df.groupby('Club',as_index=False)['Assists'].sum())
sns.set_theme(style='whitegrid',color_codes=True)
ax = sns.barplot(x='Club',y='Assists',data=Assists_by_clubs.sort_values(by="Assists"),palette='Set2')
ax.set_xlabel("Club",fontsize=30)
ax.set_ylabel("Assists",fontsize=20)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"] =(15,8)
plt.title("Plot of clubs vs Total Assist",fontsize=20)
```

```python
#top 10 assists
top10_assists = epf_df[['Name','Club','Assists','Matches']].nlargest(n=10,columns='Assists')
top10_assists
```

```python
Goals_by_club = pd.DataFrame(epf_df.groupby('Club',as_index=False)['Goals'].sum())
sns.set_theme(style='whitegrid',color_codes=True)
ax = sns.barplot(x='Club',y='Goals',data=Goals_by_club.sort_values(by='Goals'),palette='rocket')
ax.set_xlabel("Club",fontsize=30)
ax.set_ylabel("Goals",fontsize=20)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"]==(20,8)
plt.title('Plot of Clubs vs Total Goals',fontsize=20)
```

```python
#most goals by players
top_10_goals = epf_df[['Name','Club','Goals','Matches']].nlargest(n=10,columns='Goals')
top_10_goals
```

```python
#goals per match
top10_goal_per_match = epf_df[['Name','GoalsPerMatch','Matches','Goals']].nlargest(n=10, columns='GoalsPerMatch')
top10_goal_per_match
```

```python
#pie-chart goals with assist and without assist
plt.figure(figsize=(14,7))
assists = epf_df['Assists'].sum()
data = [Total_Goals - assists, assists]
labels = ['Goal without Assist','Goal with Assist']
color = sns.color_palette('Set1')
plt.pie(data,labels=labels, colors= color, autopct='%.0f%%')
plt.show()
```

```python
#Top 10 Players with most yellow cards
epf_yellow=epf_df.sort_values(by='Yellow_Cards',ascending=False)[:10]
plt.figure(figsize=(20,6))
plt.title("Players with most Yellow Cards")
c=sns.barplot(x=epf_yellow['Name'],y=epf_yellow['Yellow_Cards'],label='Players',color='Yellow')
plt.ylabel('Number of Yellow Cards')
c.set_xticklabels(c.get_xticklabels(),rotation=45)
c
```
