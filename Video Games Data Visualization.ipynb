# Data Visualization

```python
plt.figure(figsize=(10,8))
sns.barplot(data=df,x='platform',y='year')
plt.title('Platform Vs Year')
plt.xlabel('Platform')
plt.ylabel('Year')
plt.xticks(rotation=45)
plt.show()
```
