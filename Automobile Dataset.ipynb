## Handling missing values

```python
nl_mean = auto['normalized-losses'].astype(float).mean()
auto['normalized-losses'] = auto['normalized-losses'].fillna(nl_mean)

#impute data on number of doors
auto['num-of-doors']= auto['num-of-doors'].fillna("four")

#impute data on bore
bore_mean = auto['bore'].astype(float).mean()
auto['bore']= auto['bore'].fillna(bore_mean).astype(float)

#impute data on stroke
stroke_mean = auto['stroke'].astype(float).mean()
auto['stroke'] = auto['stroke'].fillna(stroke_mean).astype(float)

#imput data onhorsepower
horse_mean = auto['horsepower'].astype(float).mean()
auto['horsepower']=auto['horsepower'].fillna(horse_mean).astype(float)

#imput data on peak-rpm
peak_mean = auto['peak-rpm'].astype(float).mean()
auto['peak-rpm']=auto['peak-rpm'].fillna(peak_mean).astype(float)

#imput data on price
price_mean = auto['price'].astype(float).mean()
auto['price']=auto['price'].fillna(price_mean).astype(float)

print()
```

```python
sns.heatmap(data=auto.isna(),cmap='viridis')
```

Filled in the missing values with column means, and modified to digit types.
