# Data Visualization

```python
plt.plot(iris.sepal_length,iris.sepal_width)
plt.title("Sepal Information")
plt.xlabel("Sepal_length")
plt.ylabel("Sepal_width")
```

```python
plt.plot(iris.petal_length,iris.petal_width)
plt.title("Petal Details")
plt.xlabel("Petal length")
plt.ylabel("Petal width")
plt.show()
```

```python
plt.bar(iris.species,iris.petal_length)
plt.xlabel("Species")
plt.ylabel("petal length")
plt.title("Varities in species")
plt.show()
```

```python
plt.bar(iris.species,iris.petal_width)
plt.xlabel("Species")
plt.ylabel("Petal width")
plt.title("Petal width vs species")
plt.show()
```

```python
plt.bar(iris.species,iris.sepal_length)
plt.xlabel("Species")
plt.ylabel("sepal length")
plt.title("Sepal length Vs Species")
plt.show()
```

```python
plt.bar(iris.species,iris.sepal_width)
plt.xlabel("Species")
plt.ylabel("Sepal Width")
plt.title('Sepal width Vs Species')
plt.show()
```
