```python
#loading dependencies
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
#loading dataset to pandas dataframe
ozon = pd.read_csv(r"C:\Users\User\Downloads\Ozone Layer\eighthr.data.csv",header=None)
```

```python
#finding dtype in dataset
ozon.info()
```

```python
#finding null values
ozon.isnull().sum()
```

```python
ozon = ozon.drop([0],axis=1)
```

```python
for i in ozon.columns:
    ozon[i]=ozon[i].replace(['?'],np.nan)
```

```python
ozon.head()
```

```python
# showing column wise %ge of NaN values they contains
null_col =[]

for i in ozon.columns:
    print(i,"\t-\t",ozon[i].isnull().mean()*100)
    if ozon[i].isna().mean()*100>0:
        null_col.append(i)
```

Since no column has signficant missing values, there is no need to drop column here . Now fill the num values the mean of columns

```python
# Since dataset contains values in str format, changing it into float values
for i in ozon.columns[:-1]:
    ozon[i]=ozon[i].astype(str).astype(float)
```

```python
for i in null_col:
    ozon[i]=ozon[i].fillna(ozon[i].mean())
    
#lets check the null values again
for i in ozon.columns:
    print(i,"\t-\t",ozon[i].isna().mean()*100)
```

```python
# Checking for unbalanced dataset
plt.figure(figsize=(5,5))
ax = sns.countplot(x=73,data=ozon)

for p in ax.patches:
    ax.annotate('{}'.format(p.get_height()),(p.get_x()+0.1,p.get_height()+50))
```

```python
from imblearn.over_sampling import RandomOverSampler

oversample = RandomOverSampler()
x,y = oversample.fit_resample(ozon.drop([73],axis=1),ozon[73])

new_df = pd.DataFrame(x,columns=ozon.drop([73],axis=1).columns)
new_df[73]=y

new_df.head()
```
